```python
import numpy as np
import matplotlib.pyplot as plt

N = 500 
M = 500 

k_values1 = []
k_values2 = []

for a in range(int((N-1)/2)):
    for b in range(int((N-1)/2)):
        rhs = N * np.arccos(0.5 * (np.cos(2 * np.pi * a / N) + np.cos(2 * np.pi * b / N)))
        k_values1.append(rhs)

for m in range(M-1):
    for n in range(M-1):
        k_values2.append(np.sqrt(2) * np.pi * np.sqrt(m**2 + n**2))

all_k_values = k_values1 + k_values2
k_min = min(all_k_values)
k_max = max(all_k_values)
num_bins = 200  
bin_edges = np.linspace(k_min, k_max, num_bins + 1)

plt.figure(figsize=(10, 6))
plt.hist(k_values1, bins=bin_edges, color='blue', alpha=0.5, edgecolor='black', label='Equation 1')
plt.hist(k_values2, bins=bin_edges, color='green', alpha=0.5, edgecolor='black', label='Equation 2')
plt.title('Histogram of k values for Equations 1 and 2')
plt.xlabel('k value')
plt.ylabel('Frequency')
plt.legend()
plt.xlim(0, 1000)
plt.ylim(0, 2000)
plt.show()
```
