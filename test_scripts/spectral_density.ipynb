```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k_approx = lambda N, m, n, c: N * (np.arccos(1 / 2 * (np.cos(2 * np.pi * m / N) + np.cos(2 * np.pi * n / N))) + 2 * np.pi * c)
k_exact = lambda m, n: np.sqrt(((2 * m * np.pi)**2 + (2 * n * np.pi)**2)/2)
```

```python
N_exact = 1300
k_values2 = []
for m in range(N_exact):
    for n in range(N_exact):
        k_values2.append(k_exact(m, n))
```

```python
N_approx = 300 

k_values1 = []

for m in range(int((N_approx - 1) / 2)):
    for n in range(int((N_approx - 1) / 2)):
        for c in range(4):
            k_values1.append(k_approx(N_approx, m, n, c))

all_k_values = k_values1 + k_values2
k_min = min(all_k_values)
k_max = max(all_k_values)
num_bins = 200  
bin_edges = np.linspace(k_min, k_max, num_bins + 1)

plt.figure(figsize=(10, 6))
plt.hist(k_values1, bins=bin_edges, color='blue', alpha=0.5, edgecolor='black')
# plt.hist(k_values2, bins=bin_edges, color='green', alpha=0.5, edgecolor='black', label='Equation 2')

plt.xticks(fontsize=14, fontweight='bold')
plt.yticks(fontsize=14, fontweight='bold')

plt.xlim(0, 6000)
# plt.ylim(0, 6000)
# plt.show()

plt.savefig("square_grid_spectral_density.png", dpi=200, bbox_inches="tight")
```

```python
N = 400 
M = 500 

k_values1 = []
k_values2 = []

for m in range(int((N-1)/2)):
    for n in range(int((N-1)/2)):
        for c in range(3):
            rhs = k_approx(m, n, c)
            k_values1.append(rhs)

for m in range(M):
    for n in range(M):
        k_values2.append(k_exact(m, n))

all_k_values = k_values1 + k_values2
k_min = min(all_k_values)
k_max = max(all_k_values)
num_bins = 300  
bin_edges = np.linspace(k_min, k_max, num_bins + 1)

plt.figure(figsize=(10, 6))
plt.hist(k_values1, bins=bin_edges, color='blue', alpha=0.5, edgecolor='black', label='Equation 1')
plt.hist(k_values2, bins=bin_edges, color='green', alpha=0.5, edgecolor='black', label='Equation 2')
plt.title('Histogram of k values for Equations 1 and 2')
plt.xlabel('k value')
plt.ylabel('Frequency')
plt.legend()
plt.xlim(0, 1000)
plt.ylim(0, 2000)
plt.show()
```
